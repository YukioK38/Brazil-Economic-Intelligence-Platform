```python
import pandas as pd
import os
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()
engine = create_engine(os.getenv("DATABASE_URL"))
```

```python
# Queries
query_database_schema = """
    SELECT * 
    FROM information_schema.tables 
    WHERE table_schema = 'public';
"""
query_head="""
    SELECT *
    FROM {}
    LIMIT 10;
"""

query_join_indicators="""
    SELECT *
    FROM indicator_values as v
    INNER JOIN indicators as i
        ON v.indicator_id = i.indicator_id;
"""

query_join_derived_indicators="""
    SELECT *
    FROM derived_values as v
    INNER JOIN derived_indicators as i
        ON v.derived_id = i.derived_id
"""
```

```python
pd.read_sql_query(query_database_schema, engine)
```

```python
table_names = ["indicator_values", "indicators", "derived_values", "derived_indicators"]
for name in table_names:
    query = query_head.format(name)
    print(f"Table name:{name}")
    display(pd.read_sql_query(query, engine))
    print()
```

```python
df_indicators = pd.read_sql_query(query_join_indicators, engine)
df_derived = pd.read_sql_query(query_join_derived_indicators, engine)
print("Joined indicators table head: ")
print(df_indicators.head())
print()
print("Joined derived indicators table head: ")
print(df_derived.head())
```
